```python
import pandas as pd
import ivolatility as ivol
from dotenv import load_dotenv
import os

# Adjust pandas settings to display all columns
pd.set_option('display.max_columns', None)

# Load environment variables from .env file
load_dotenv()

# Retrieve the API key from the .env file
api_key = os.getenv("IVOLATILITY_API_KEY")

# Set login parameters for ivolatility
ivol.setLoginParams(apiKey=api_key)
getMarketData = ivol.setMethod('/equities/eod/stock-opts-by-param')
marketData = getMarketData(
    symbol='VXX',
    tradeDate='2021-12-16',
    dteFrom=14,
    dteTo=360,
    deltaFrom=0,
    deltaTo=1,
    cp='C',
    region="USA"
    )

df = pd.DataFrame(marketData)
df.head(50)
```
